```python
def ConvertRange(x):
    temp = x.split('-')
    if len(temp) == 2:
        return (float(temp[0]) + float(temp[1]))/2
    try:
        return float(x)
    except:
        return None
```

```python
def remove_outliers_sqft(df):
    df_output = pd.DataFrame()
    for key,subdf in df.groupby(['location', 'area_type']):
        m = np.mean(subdf.price_per_sqft)
        st = np.std(subdf.price_per_sqft)
        gen_df = subdf[(subdf.price_per_sqft > (m-st))&(subdf.price_per_sqft < (m+st))]
        df_output = pd.concat([df_output, gen_df], ignore_index = True)
    return df_output
data = remove_outliers_sqft(data)
data.describe()
```

```python
data.shape
```

```python
def remove_outliers_bhk(df):
    exclude_indices = np.array([])
    for loc, locdf in df.groupby(['location', 'area_type']):
        bhk_stats = {}
        for bhk, bhk_df in locdf.groupby('BHK'):
            bhk_stats[bhk] = {
                'mean' : np.mean(bhk_df.price_per_sqft),
                'std' : np.mean(bhk_df.price_per_sqft),
                'count' : bhk_df.shape[0]
            }
        for bhk, bhk_df in locdf.groupby('BHK'):
            stats = bhk_stats.get(bhk - 1)
            if stats and stats['count']>5:
                exclude_indices = np.append(exclude_indices, bhk.df[bhk_df.price_per_sqft<(stats['mean'])].index.values)
        return df.drop(exclude_indices, axis = 'index')
```

```python
data = remove_outliers_bhk(data)
```

```python
X = data.drop(columns = ['price'])
y = data['price']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, Lasso, Ridge
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
from sklearn.metrics import r2_score
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.2, random_state = 0)
```

```python
print(X_train.shape)
print(X_test.shape)
```

Linear Regression

```python
column_trans = make_column_transformer((OneHotEncoder(sparse_output = False),['location', 'area_type']),
                                      remainder='passthrough')
```

```python
scaler = StandardScaler()
```

```python
lr = LinearRegression()
```

```python
pipe = make_pipeline(column_trans, scaler, lr)
```

```python
pipe.fit(X_train, y_train)
```

```python
y_pred_lr = pipe.predict(X_test)
```

```python
r2_score(y_test, y_pred_lr)
```

Lasso

```python
lasso = Lasso()
```

```python
pipe = make_pipeline(column_trans, scaler, lasso)
```

```python
pipe.fit(X_train, y_train)
```

```python
y_pred_lasso = pipe.predict(X_test)
```

```python
r2_score(y_test, y_pred_lasso)
```

Ridge

```python
ridge = Ridge()
```

```python
pipe = make_pipeline(column_trans, scaler, ridge)
```

```python
pipe.fit(X_train, y_train)
```

```python
y_pred_ridge = pipe.predict(X_test)
```

```python
r2_score(y_test, y_pred_ridge)
```
